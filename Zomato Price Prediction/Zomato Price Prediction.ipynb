# Handle rate col

```python
df['rate'].unique()
```

```python
def handrate(V):
    if (V == 'NEW' or V =='-'):
        return np.nan
    else:
        V = str(V).split('/')[0]
        return float(V)
```

```python
df['rate'] = df['rate'].apply(handrate)
df['rate'].head()
```

```python
df['rate'].unique()
```

```python
df['rate'].isna().sum()
```

```python
df['rate'].fillna(df['rate'].median(), inplace=True)
df['rate'].isna().sum()
```

# Handle comma col

```python
def handcomma(v):
    v = str(v)
    if ',' in v:
        v=v.replace(',','')
        return float(v)
    else:
        return float(v)
```

```python
df['approx_cost(for two people)'] = df['approx_cost(for two people)'].apply(handcomma)
df['approx_cost(for two people)'].unique()
```

```python
df['approx_cost(for two people)'].isna().sum()
```

# Handle rest_type col

```python
df['rest_type'].value_counts()
```

```python
rest_type =df['rest_type'].value_counts(ascending=False)
rest_type
```

```python
other = rest_type[rest_type<1000]
other
```

```python
def handle_rest_type(value):
    if value in other:
        return 'other'
    else:
        return value
```

```python
df['rest_type'] = df['rest_type'].apply(handle_rest_type)
df['rest_type'].value_counts()
```

# Handle Location col

```python
location = df['location'].value_counts(ascending=False)
loc300 = location[location<390]
```

```python
def handloc(v):
    if(v in loc300 ):
        return 'other'
    else:
        return v
```

```python
df['location'] =df['location'].apply(handloc)
df['location'].value_counts(ascending=False)
```

# Handle cuisines col

```python
cuisines = df['cuisines'].value_counts(ascending=False)
cuisines100 = cuisines[cuisines<100]
```

```python
def handcusi(v):
    if (v in cuisines100):
        return 'other'
    else:
        return v
```

```python
df['cuisines'] = df['cuisines'].apply(handcusi)
df['cuisines'].value_counts(ascending=False)
```

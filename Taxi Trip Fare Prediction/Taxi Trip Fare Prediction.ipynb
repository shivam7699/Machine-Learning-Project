```python
# Cleaning store_and_fwd_flag col
```

```python
df_train['store_and_fwd_flag'].unique()
```

```python
def hand_store_and_fwd_flag(v):
    if (v=='N'):
        return 0
    else:
        return 1        
```

```python
df_train['store_and_fwd_flag'] = df_train['store_and_fwd_flag'].apply(hand_store_and_fwd_flag)
df_test['store_and_fwd_flag'] = df_test['store_and_fwd_flag'].apply(hand_store_and_fwd_flag)
```

```python
from sklearn.preprocessing import MinMaxScaler
Scaler = MinMaxScaler()
```

```python
final_train_cols = list(df_train.columns)
final_train_cols.remove('calculated_total_amount')
final_train_cols
```

```python
Scaler.fit(df_train[final_train_cols])
```

```python
df_train[final_train_cols] = Scaler.transform(df_train[final_train_cols])
```

```python
final_test_cols = list(df_test.columns)
```

```python
Scaler.fit(df_test[final_test_cols])
```

```python
df_test[final_test_cols] = Scaler.transform(df_test[final_test_cols])
```

```python
#model bulding
```

```python
X_train = df_train[final_train_cols]
y_train = df_train['calculated_total_amount']
```

```python
# 
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(X_train,y_train)
model.score(X_train,y_train)
```

```python
# Random Forest 
from sklearn.ensemble import RandomForestRegressor
model = RandomForestRegressor()
model.fit(X_train,y_train)
model.score(X_train,y_train)
```

```python
# Decision Tree
from sklearn.tree import DecisionTreeRegressor
model = DecisionTreeRegressor()
model.fit(X_train,y_train)
model.score(X_train,y_train)
```
